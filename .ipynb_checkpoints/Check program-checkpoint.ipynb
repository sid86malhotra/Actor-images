```python
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
import matplotlib.pyplot as plt

def cross_entropy_loss(Y, T):
    N = len(T)
    return -(np.log(Y[np.arange(N), T])).sum()

def calcualte_accuracy(output, Y):
    pY = np.argmax(output, axis = 1)
    return np.mean(pY == Y)

#Now we convert the target values to numeric
def label_onehot_encode(Y):

    N = len(Y)

    output = []
    T = np.zeros((N, len(set(Y))))

    for row in range(N):
        if Y[row] == "OLD":
            T[row, 2] = 1
            output.append(2)
        elif Y[row] == "MIDDLE":
            T[row, 1] = 1
            output.append(1)
        else:
            T[row, 0] = 1
            output.append(0)

    outputNUMPY = np.array(output)
    return T, outputNUMPY


class FFNNUMPY(object):

    def __init__(self, M):
        self.M = M

    def predict(self, X, get_weights = False):
        if get_weights:
            self.W1 = pd.read_csv("W1.csv").as_matrix()
            self.W2 = pd.read_csv("W2.csv").as_matrix()
            self.b1 = pd.read_csv("b1.csv").as_matrix()
            self.b2 = pd.read_csv("b2.csv").as_matrix()
        hidden = np.tanh(X.dot(self.W1) + self.b1)
        output = np.softmax(hidden.dot(self.W2) + self.b2)
        return hidden, output

    def initiate_weights(self, N, D, K):
        W1 = np.random.randn(D, self.M) / np.sqrt(N + D)
        b1 = np.zeros(self.M)
        W2 = np.random.randn(self.M, K) / np.sqrt(N + D)
        b2 = np.zeros(K)
        print(W1, b1, W2, b2)
        return W1, b1, W2, b2

    def save_weights_func():

        df = pd.Dataframe(self.W1)
        df.to_csv("W1.csv")
        df = pd.Dataframe(self.W2)
        df.to_csv("W2.csv")
        df = pd.Dataframe(self.b1)
        df.to_csv("b1.csv")
        df = pd.Dataframe(self.b2)
        df.to_csv("b2.csv")

    def score_function(self, X, Y, get_weights = False):
        _, output = predict(self, X, get_weights)
        return calcualte_accuracy(output, Y)

    def fit(self, X, Y, T, learning_rate = 10e-7, reg = 10e-7, epochs = 20, batch_size = 500, show_fig = True, save_weights = True):

        N, D = X.shape
        K = len(set(Y))
        
        self.W1, self.b1, self.W2, self.b2 = self.initiate_weights(N, D, K)
        
        print(X.shape, self.W1.shape)
        
        num_batches = np.round(N / batch_size)

        costs = []

        for epoch in range(epochs):
            X, Y, T = shuffle(X, Y, T)

            for batch in range(batch_size):

                Xbatch = X[batch * batch_size: batch_size * (batch + 1)]
                Tbatch = T[batch * batch_size: batch_size * (batch + 1)]

                hidden, output = self.predict(Xbatch)

                pY_T = (output - Tbatch)

                W2 -= learning_rate * (hidden.dot(pY_T) + reg * self.W2)
                b2 -= learning_rate * (pY_T.sum(axis = 0) + reg * self.b2)

                dZ = pY_T.dot(W2.T) * hidden * (1 - hidden)

                W1 -= learning_rate * (Xbatch.T.dot(dZ) + reg * self.W1)
                b1 -= learning_rate * (dZ.sum(axis = 0) + reg * self.b2)

            hidden, output = self.predict(X)
            c = cross_entropy_loss(output, T)
            a = calcualte_accuracy(output, Y)
            costs.append(c)
            print("Epoch", epoch, "Costs", c, "Accuracy", a)

        if show_fig:
            _ = plt.plot(costs)
            plt.show()
```

```python

data = pd.read_csv("Training.csv").as_matrix()

X = (data[:, 2:] / 255).astype(np.float32)
Y = data[:,1]

T, Y = label_onehot_encode(Y)

#We would need to setup the training and testing samples from the data
X, T = shuffle(X, T)
```

```python

Xtrain, Ytrain, Ttrain = X[:-1000], Y[:-1000], T[:-1000]
Xtest, Ytest, Ttest = X[-1000:], Y[-1000:], T[-1000:]
```

```python
model = FFNNUMPY(M = 200)

model.fit(Xtrain, Ytrain, Ttrain)
model.score(Xtest, Ytest)
```
